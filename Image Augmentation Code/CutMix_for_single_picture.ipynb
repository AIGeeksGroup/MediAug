```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random

# Load the image
image_path = '/content/drive/My Drive/Eye_rgb/1144_left.jpg'
image = cv2.imread(image_path)
if image is None:
    raise ValueError("Image not found. Check the file path.")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for visualization

# Get the image size
H, W, C = image.shape

# Define the size of the region to cut (1/4 of the image)
cut_h = H // 2  # Half the height
cut_w = W // 2  # Half the width

# Randomly select the top-left corner of the cut region
x1 = random.randint(0, W - cut_w)
y1 = random.randint(0, H - cut_h)
x2 = x1 + cut_w
y2 = y1 + cut_h

# Cut the region
cut_region = image[y1:y2, x1:x2].copy()

# Randomly select the position to paste the cut region
paste_x1 = random.randint(0, W - cut_w)
paste_y1 = random.randint(0, H - cut_h)
paste_x2 = paste_x1 + cut_w
paste_y2 = paste_y1 + cut_h

# Create a copy of the original image to modify
modified_image = image.copy()

# Paste the cut region onto the new location
modified_image[paste_y1:paste_y2, paste_x1:paste_x2] = cut_region

# Plot the original and modified images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(modified_image)
plt.title("Modified Image (Cut and Moved)")
plt.axis("off")

plt.show()
```
